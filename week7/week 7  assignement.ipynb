##### Q1--4

```python

#q1
from random import random, seed, sample



randlist=[]
for i in range (1000000):
    randlist.append(random())
    


#q2
print("First 10 numbers:", randlist[:10])
print("Last 10 numbers:", randlist[-10:])
print("Maximum number:", max(randlist))
print("Minimum number:", min(randlist))

#q3
sublist1 = randlist[::10000]

print ("Sublist 1", sublist1)

# q4
sublist2 = sample(randlist, 100)

print("Sublist 2:", sublist2)



```

###### Q5

```python
#q5
import statistics as stat

means1 = statistics.mean(sublist1)
print ("mean of sublsit1:", means1)
means2 = statistics.mean(sublist2)
print ("mean of sublsit2:", means2)

stdvs1= stat.stdev(sublist1)
print ("Standard deviation of sublist 1:", stdvs1)
stdvs2= stat.stdev(sublist2)
print ("Standard Deviation of sublist 1:", stdvs2)

var1= stat.variance(sublist1)
print("Variance of sublist 1:",var1)
var2= stat.variance(sublist2)
print("Variance of sublsit 2:",var2)
```

##### Q6

```python
import math as m

sublist=[1,4,6,56,5,76,8,679,786,454,326568,79,675,63523,564,378567,9,678657,634,56,8,7,32,13432,5678,52124,3567524345,672431,2435,6724,3112,34567,6524,31,2345,6786,52431,23456,78967,52431,2435678]

print ("Statisticts of first sample \n\n")

mean=m.fsum(sublist)/len(sublist)
print("Mean (using math module):{}".format(mean))

mylist=[]
for i in sublist:
    mylist.append((i-mean)**2)

variance=m.fsum(mylist)/(len(sublist)-1)

print ("Variance (using math module):{}".format(variance))

sd=m.sqrt(variance)

print("Standard deviation (using math module):{}".format(sd))
```
